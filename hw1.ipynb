## 问题 3

给定一个包含n个元素的数组 `profits` , 它的第 `i` 个元素 `profits[i]` 表示一支股票第 `i` 天的**收益**（正数表示涨, 负数表示跌）。你只能选择 **某一天** 买入这只股票, 并选择在 **未来的某一个不同的日子** 卖出该股票。

1. 设计一个算法来计算你所能获取的最大利润和对应买入和卖出的日期。请分析算法方案, 计算其时间复杂度, 并且使用python编程实现该算法。

2. \* 设计一个时间复杂度为 $O(n)$的算法实现该算法

e.g. :
---
profits=[3,2,1,-7,5,2,-1,3,-1], 第5天买入, 第8天卖出, 收益最大：9

idea:

````python
# add your code here
def max_profit_days(profits):
    n = len(profits)
    if n < 2:
        return None, -1, -1
    cur_sum = profits[0] + profits[1]
    best_sum = cur_sum
    cur_start = 0
    best_start = 0
    best_end = 1
    for i in range(2, n):
        if cur_sum + profits[i] < profits[i - 1] + profits[i]:
            cur_sum = profits[i - 1] + profits[i]
            cur_start = i - 1
        else:
            cur_sum += profits[i]
        if cur_sum > best_sum:
            best_sum = cur_sum
            best_start = cur_start
            best_end = i
    return best_sum, best_start + 1, best_end + 1
profits = [3, 2, 1, -7, 5, 2, -1, 3, -1]
max_profit, buy_day, sell_day = max_profit_days(profits)
print("最大收益:", max_profit)
print("买入日期:", buy_day)
print("卖出日期:", sell_day)
# your algorithm time complexity is: O(n)
````

给定一个包含不同整数元素的数组 $ A[1..n]$ ,并且满足条件：$A[1]>A[2]$ 并且 $A[n-1]<A[n]$; 规定：如果一个元素比它两边的邻居元素都小, 即：$A[x]<A[x-1], A[x]<A[x+1]$ , 称这个元素A[x]为“局部最小”。通过遍历一次数组, 我们可以很容易在 $O(n)$的时间复杂度下找到一个局部最小值, 


1. 分析该问题, 设计一个算法在$O(logn)$的时间复杂度下找到一个局部最小(返回数值), 要求：分析算法设计思路, 并且使用python编程实现
2. \* 设计算法找出所有局部最小值, 分析算法设计思路, 并使用python编程实现

e.g.:
---
A=[9, 3, 7, 2, 1, 4, 5 ] 时,  局部最小元素为 3, 1

idea:

````python
# add your code here
def find_local_minimum(A):
    n = len(A)
    left = 1
    right = n - 2
    while left <= right:
        mid = (left + right) // 2
        if A[mid] < A[mid - 1] and A[mid] < A[mid + 1]:
            return A[mid]
        elif A[mid] > A[mid - 1]:
            right = mid - 1
        else:
            left = mid + 1
    return None
A = [9, 3, 7, 2, 1, 4, 5]
ans = find_local_minimum(A)
print("找到的局部最小值为:", ans)
# your algorithm time complexity is: O(logn)
````

e.g.：
---
数组arr=[5,4,3,2,6,1,88,33,22,107] 的前3个最小数据为：1, 2, 3

idea:

````python
# add your code here
def quick_order(nums):
    if len(nums) <= 1:
        return nums[:]
    base = nums[len(nums) // 2]
    smaller = []
    same = []
    bigger = []
    for value in nums:
        if value < base:
            smaller.append(value)
        elif value > base:
            bigger.append(value)
        else:
            same.append(value)
    return quick_order(smaller) + same + quick_order(bigger)
def get_by_sort(sequence, need):
    ordered = quick_order(sequence)
    answer = []
    for i in range(need):
        answer.append(ordered[i])
    return answer
class SmallRootTree:
    def __init__(self):
        self.box = []
    def push(self, number):
        self.box.append(number)
        cur = len(self.box) - 1
        while cur > 0:
            up = (cur - 1) // 2
            if self.box[up] <= self.box[cur]:
                break
            self.box[up], self.box[cur] = self.box[cur], self.box[up]
            cur = up
    def pop_smallest(self):
        top = self.box[0]
        tail = self.box.pop()
        if self.box:
            self.box[0] = tail
            cur = 0
            size = len(self.box)
            while True:
                left = cur * 2 + 1
                right = cur * 2 + 2
                choose = cur
                if left < size and self.box[left] < self.box[choose]:
                    choose = left
                if right < size and self.box[right] < self.box[choose]:
                    choose = right
                if choose == cur:
                    break
                self.box[cur], self.box[choose] = self.box[choose], self.box[cur]
                cur = choose
        return top
def get_by_heap(sequence, need):
    tree = SmallRootTree()
    for number in sequence:
        tree.push(number)
    answer = []
    for i in range(need):
        answer.append(tree.pop_smallest())
    return answer
arr = [5, 4, 3, 2, 6, 1, 88, 33, 22, 107]
k = 3
result1 = get_by_sort(arr, k)
print("方法一结果：", result1)
result2 = get_by_heap(arr, k)
print("方法二结果：", result2)
# your algorithm time complexity is: 方法一：O(nlogn)；方法二：O((n+k)logn)
````

>
> $ A=[23,-8, 9, -3, -7, 9, -23, 22] $

请设计一个算法模拟星球的运行情况，输出最终的星球存续情况（输出一个数组），分析算法设计思路，计算时间复杂度，并基于python编程实现。

e.g.
---
1.  输入： A=[-3,-6,2,8, 5,-8,9,-2,1]， 输出：[-3, -6, 2, 8, 9, 1]

2. 输入：A=[23,-8, 9, -3, -7, 9, -23, 22], 输出：[23, 22]

idea:

````python
# add your code here
def star_result(stars):
    keep = []
    for cur in stars:
        alive = True
        while alive and keep and keep[-1] > 0 and cur < 0:
            top = keep[-1]
            if abs(top) < abs(cur):
                keep.pop()
            else:
                alive = False
        if alive:
            keep.append(cur)
    return keep
arr1 = [-3, -6, 2, 8, 5, -8, 9, -2, 1]
arr2 = [23, -8, 9, -3, -7, 9, -23, 22]
print("样例1输出:", star_result(arr1))
print("样例2输出:", star_result(arr2))
# your algorithm time complexity is: O(n)
````

## 问题 7

> 0-1背包问题：现在有4块大理石，每个大理石对应的重量和价值使用一个元组表示, 即（重量，价值），4块大理石的重量和价值为：[(5,10), (4,40), (6,30), (3, 50)], 假设你有一辆最大承重为9的小推车，请问使用这个小推车装大理石的最大价值为多少

请设计一个算法解决该问题，分析算法设计思路，计算时间复杂度，并基于python编程实现

idea:

````python
# add your code here
def get_max_value(items, max_weight):
    value_table = [0] * (max_weight + 1)
    for stone_weight, stone_value in items:
        for cur_weight in range(max_weight, stone_weight - 1, -1):
            new_value = value_table[cur_weight - stone_weight] + stone_value
            if new_value > value_table[cur_weight]:
                value_table[cur_weight] = new_value
    return value_table[max_weight]
marbles = [(5, 10), (4, 40), (6, 30), (3, 50)]
limit = 9
result = get_max_value(marbles, limit)
print("最大价值为:", result)
# your algorithm time complexity is: O(nC)，其中大理石数量为 n，小推车最大承重为 C
````

e.g.:

输入：root = [1,3,null,null,2]， 输出为：[3,1,null,null,2]， 

解释：3 不能是 1 的左孩子，因为 3 > 1 。交换 1 和 3 使二叉搜索树有效。

2.

 <div style="text-align:center;width:200px;"> <img alt="" src="./fig/hw3q9-1.jpg"> </div>
输入： root=[3,1,4,null,null,2]， 输出为：[2,1,4,null,null,3]

解释：2 不能在 3 的右子树中，因为 2 < 3 。交换 2 和 3 使二叉搜索树有效。

idea:

````python
# add your code here
from collections import deque
class Node:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None
def make_tree(values):
    if not values or values[0] is None:
        return None
    root = Node(values[0])
    waiting = deque([root])
    index = 1
    while waiting and index < len(values):
        cur = waiting.popleft()
        if index < len(values):
            left_value = values[index]
            index += 1
            if left_value is not None:
                cur.left = Node(left_value)
                waiting.append(cur.left)
        if index < len(values):
            right_value = values[index]
            index += 1
            if right_value is not None:
                cur.right = Node(right_value)
                waiting.append(cur.right)
    return root
def output_tree(root):
    if root is None:
        return []
    q = deque([root])
    ans = []
    while q:
        cur = q.popleft()
        if cur is None:
            ans.append(None)
        else:
            ans.append(cur.val)
            q.append(cur.left)
            q.append(cur.right)
    while ans and ans[-1] is None:
        ans.pop()
    return ans
def fix_search_tree(root):
    stack = []
    cur = root
    previous = None
    first_wrong = None
    second_wrong = None
    while stack or cur:
        while cur:
            stack.append(cur)
            cur = cur.left
        cur = stack.pop()
        if previous is not None and previous.val > cur.val:
            if first_wrong is None:
                first_wrong = previous
            second_wrong = cur
        previous = cur
        cur = cur.right
    if first_wrong is not None and second_wrong is not None:
        first_wrong.val, second_wrong.val = second_wrong.val, first_wrong.val
    return root
test1 = [1, 3, None, None, 2]
tree1 = make_tree(test1)
fix_search_tree(tree1)
print("样例1输出:", output_tree(tree1))
test2 = [3, 1, 4, None, None, 2]
tree2 = make_tree(test2)
fix_search_tree(tree2)
print("样例2输出:", output_tree(tree2))

# your algorithm time complexity is: O(n)
````

给定 `n` 个非负整数表示每个宽度为 `1` 的柱子的高度图，计算按此排列的柱子，下雨之后能接多少雨水。

请分析算法设计思路，计算时间复杂度，并基于python编程实现

idea:

````python
# add your code here
def rain_amount(walls):
    saved = []
    water = 0
    for pos, high in enumerate(walls):
        while saved and high > walls[saved[-1]]:
            bottom = saved.pop()
            if not saved:
                break
            left_wall = saved[-1]
            width = pos - left_wall - 1
            bounded_height = min(walls[left_wall], high) - walls[bottom]
            water += width * bounded_height
        saved.append(pos)
    return water
height1 = [0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1]
height2 = [4, 2, 0, 3, 2, 5]
print("示例1结果:", rain_amount(height1))
print("示例2结果:", rain_amount(height2))

# your algorithm time complexity is: O(n)
````

## 问题 10

给你一个二叉树的根节点 root ，树中每个节点都存放有一个 0 到 9 之间的数字。每条从根节点到叶节点的路径都代表一个数字：

> 例如，从根节点到叶节点的路径 1 -> 2 -> 3 表示数字 123 。

设计一个算法计算从根节点到叶节点生成的所有数字之和 ，分析算法设计思路，计算时间复杂度，并基于python编程实现

叶节点:是指没有子节点的节点。

e.g.:
---
1.

 <div style="text-align:center;width:200px;"> <img alt="figure" src="./fig/hw3q10-1.jpg"> </div>

 输入：root = [1,2,3], 输出：25

解释：从根到叶子节点路径 1->2 代表数字 12，1->3 代表数字 13，因此和为 12 + 13 = 25 。

2.

 <div style="text-align:center;width:200px;"> <img alt="figure" src="./fig/hw3q10-2.jpg"> </div>

 输入：root = [4,9,0,5,1], 输出：1026

解释：从根到叶子节点路径 4->9->5 代表数字 495，4->9->1 代表数字 491，4->0 代表数字 40。因此和为 495 + 491 + 40 = 1026 。

idea:

````python
# add your code here
from collections import deque
class Node:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
def build_tree(values):
    if not values or values[0] is None:
        return None
    root = Node(values[0])
    q = deque([root])
    index = 1
    while q and index < len(values):
        cur = q.popleft()
        if index < len(values) and values[index] is not None:
            cur.left = Node(values[index])
            q.append(cur.left)
        index += 1
        if index < len(values) and values[index] is not None:
            cur.right = Node(values[index])
            q.append(cur.right)
        index += 1
    return root
def sum_root_to_leaf(root):
    def dfs(node, path_value):
        if node is None:
            return 0
        new_value = path_value * 10 + node.val
        if node.left is None and node.right is None:
            return new_value
        left_sum = dfs(node.left, new_value)
        right_sum = dfs(node.right, new_value)
        return left_sum + right_sum
    return dfs(root, 0)
tree_data1 = [1, 2, 3]
tree_data2 = [4, 9, 0, 5, 1]
root1 = build_tree(tree_data1)
root2 = build_tree(tree_data2)
print("示例1结果:", sum_root_to_leaf(root1))
print("示例2结果:", sum_root_to_leaf(root2))
# your algorithm time complexity is: O(n)
````
